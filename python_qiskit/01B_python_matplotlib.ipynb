# 📊 Visualización de vectores probabilísticos con `plt.bar`

## 🔹 Objetivo
Representar un **vector de probabilidad** en un **diagrama de barras**.  
Ejemplo: una moneda cargada con probabilidades:

$$
\vec{p} = 
\begin{bmatrix}
0.3 \\ 
0.7
\end{bmatrix}
$$

donde $p_s = 0.3$ (sello) y $p_c = 0.7$ (cara).  

---

## 🔹 Sintaxis de `plt.bar`

La función principal es `plt.bar()`. Cada argumento permite configurar distintos aspectos del gráfico.

### Parámetros principales

- **`x`**  
  Posiciones o etiquetas en el eje horizontal.  
  - Ejemplo: `["Sello", "Cara"]`.

- **`height`**  
  Alturas de cada barra.  
  - En este caso: `[0.3, 0.7]` (las probabilidades).

### Parámetros opcionales de personalización

- **`width=0.8`**  
  Grosor de las barras (por defecto 0.8).  

- **`bottom=None`**  
  Punto en el eje Y desde donde arranca cada barra.  
  - Si no se especifica, arranca en `0`.  

- **`align='center'`**  
  Controla cómo se alinean las barras respecto a `x`.  
  - `'center'`: centradas en el valor.  
  - `'edge'`: alineadas por el borde.  

- **`color=None`**  
  Define los colores de las barras.  
  - Puede ser un solo color (`"blue"`) o una lista (`["skyblue", "orange"]`).  

- **`edgecolor=None`**  
  Color de los bordes de cada barra.  

- **`linewidth=None`**  
  Grosor del borde de cada barra.  

- **`tick_label=None`**  
  Etiquetas de las posiciones `x` (alternativa a pasarlas directamente como lista).  

---

## 🔹 Resumen sintáctico

```python
plt.bar(x, height, width=0.8, bottom=None, align='center', color=None, edgecolor=None, linewidth=None, tick_label=None)


````python
import numpy as np
import matplotlib.pyplot as plt

# Vector de probabilidad de la moneda [Sello, Cara]
p = np.array([0.3, 0.7])

# Etiquetas para el eje X
etiquetas = ["Sello", "Cara"]

# Crear gráfico de barras
plt.bar(
    x=etiquetas,      # categorías en el eje X
    height=p,         # altura de cada barra (las probabilidades)
    width=0.6,        # ancho de cada barra (más angostas que el valor por defecto)
    bottom=0,         # las barras parten desde 0 en el eje Y
    align='center',   # cada barra queda centrada en su categoría
    color=["skyblue", "orange"],  # colores personalizados para cada barra
    edgecolor="black", # color del borde de las barras
    linewidth=1.2      # grosor del borde
)

# Agregar título y etiquetas
plt.title("Probabilidades de la moneda")
plt.ylabel("Probabilidad")
plt.xlabel("Resultado")

# Mostrar el gráfico
plt.show()
````

# 🥧 Visualización de vectores probabilísticos con `plt.pie`

## 🔹 Objetivo
Representar un **vector de probabilidad** como un **gráfico circular** (gráfico de torta o "pie chart").  
Este tipo de gráfico permite **comparar proporciones** de manera visual.

Supongamos la misma moneda cargada con probabilidades:

$$
\vec{p} =
\begin{bmatrix}
0.3 \\
0.7
\end{bmatrix}
=
\begin{bmatrix}
p_s \\
p_c
\end{bmatrix}
$$

donde $p_s = 0.3$ (sello) y $p_c = 0.7$ (cara).  

---

## 🔹 Sintaxis de `plt.pie`

La función principal es `plt.pie()`.  

### Parámetros principales

- **`x`**  
  Lista de valores que se van a representar en proporciones.  
  - Ejemplo: `[0.3, 0.7]`.

- **`labels`** *(opcional)*  
  Etiquetas para cada porción.  
  - Ejemplo: `["Sello", "Cara"]`.

- **`colors`** *(opcional)*  
  Colores asignados a cada sector.  
  - Ejemplo: `["skyblue", "orange"]`.

- **`autopct`** *(opcional)*  
  Formato de texto que muestra el **porcentaje** en cada sector.  
  - Ejemplo: `"%.1f%%"` → mostrará un número con un decimal y el símbolo `%`.  

### Parámetros de estilo

- **`startangle=0`**  
  Ángulo inicial en grados desde donde comienza el gráfico.  
  - Ejemplo: `90` → empieza desde la parte superior.

- **`counterclock=True`**  
  Define si las porciones se dibujan en sentido antihorario (`True`) o horario (`False`).  

- **`shadow=False`**  
  Si se activa (`True`), dibuja una sombra debajo del gráfico.  

---

## 🔹 Resumen sintáctico

```python
plt.pie(x, labels=None, colors=None, autopct=None, startangle=0, counterclock=True, shadow=False)


````python
import numpy as np
import matplotlib.pyplot as plt

# Vector de probabilidad de la moneda [Sello, Cara]
p = np.array([0.3, 0.7])

# Etiquetas para cada sector del gráfico
etiquetas = ["Sello", "Cara"]

# Colores para diferenciar cada sector
colores = ["skyblue", "orange"]

# Crear gráfico de pastel
plt.pie(
    x=p,                  # valores (probabilidades)
    labels=etiquetas,     # etiquetas de cada sector
    colors=colores,       # colores personalizados
    autopct="%.1f%%",     # mostrar porcentajes con 1 decimal
    startangle=90,        # comenzar desde la parte superior
    counterclock=True,    # dibujar en sentido antihorario
    shadow=True           # activar sombra debajo del gráfico
)

# Agregar título
plt.title("Distribución de probabilidades de la moneda")

# Mostrar gráfico
plt.show()
````

# 🖼️ Uso de `plt.subplots` en Matplotlib

## 🔹 Objetivo  
`plt.subplots` permite **crear una figura con múltiples ejes (subplots) organizados en filas y columnas**.  
Esto es útil para:  

- Comparar diferentes gráficos lado a lado.  
- Visualizar distintos aspectos del mismo conjunto de datos.  
- Mantener una presentación limpia y uniforme.

---

## 🔹 Conceptos clave

1. **Figura (`fig`)**  
   - Representa el **contenedor principal** de todos los subplots.  
   - Se puede ajustar su tamaño con `figsize=(ancho, alto)`.

2. **Ejes (`axs`)**  
   - Son los **subplots individuales** dentro de la figura.  
   - Pueden ser un **array** (si hay más de uno) o un **objeto único** (si solo hay un subplot).  
   - Cada `ax` se puede usar como un eje independiente para graficar (`ax.plot(...)`, `ax.imshow(...)`, etc.).

3. **Distribución de subplots**  
   - Definida por **filas y columnas**: `plt.subplots(nrows, ncols)`.  
   - Ejemplo: `plt.subplots(2, 3)` → 2 filas y 3 columnas (6 subplots).  
   - `axs` tendrá la forma de un **array 2D**: `axs[fila, columna]`.

---

## 🔹 Pasos didácticos para usar `plt.subplots`

1. **Crear la figura y los ejes**  
   - `fig, axs = plt.subplots(1, 3, figsize=(12, 4))`  
   - Significa **1 fila y 3 columnas**, cada subplot se guarda en `axs[0]`, `axs[1]`, `axs[2]`.

2. **Graficar en cada subplot**  
   - Cada eje es independiente:  
     ```text
     axs[0].imshow(...)  # Primer subplot
     axs[1].imshow(...)  # Segundo subplot
     axs[2].imshow(...)  # Tercer subplot
     ```
   - Se pueden asignar títulos independientes a cada subplot: `axs[i].set_title("...")`.

3. **Agregar colorbars u otras anotaciones**  
   - Se pueden vincular a un subplot específico:  
     `plt.colorbar(im, ax=axs[i])`.

4. **Ajustar espaciado**  
   - `plt.tight_layout()` para que los subplots no se superpongan y se vea limpio.

5. **Mostrar la figura**  
   - `plt.show()` renderiza todos los subplots simultáneamente.

---

## 🔹 Notas pedagógicas

- Cuando se usan **muchos subplots**, conviene que la figura sea más ancha o más alta (`figsize`) para que no queden apretados.  
- Si hay un solo subplot, `axs` puede ser un objeto único en lugar de un array.  
- Se pueden usar combinaciones de `plt.subplot()` (una sola celda) para gráficos rápidos, pero `plt.subplots()` es más flexible para figuras complejas.

---

## 🔹 Sintaxis de `plt.subplots` (al final)

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(ancho, alto))


# 🌡️ Visualización de vectores probabilísticos con `plt.imshow`

## 🔹 Objetivo
Un **heatmap (mapa de calor)** es una representación donde los valores numéricos se muestran como **colores**.  
Esto permite visualizar la magnitud de las probabilidades de un vector de forma más intuitiva.

Por ejemplo, para la moneda con probabilidades:

$$
\vec{p} =
\begin{bmatrix}
0.3 \\
0.7
\end{bmatrix}
$$

podemos representar estos valores en una escala de colores.

---

## 🔹 Sintaxis de `plt.imshow`

La función es:

### Parámetros principales

- **`X`**  
  Matriz (2D) o vector convertido en matriz, cuyos valores serán representados como colores.  
  - Ejemplo: `[[0.3, 0.7]]` para visualizar un vector fila.

- **`cmap`** *(opcional)*  
  Mapa de colores usado para la escala.  
  Existen distintos esquemas que conviene elegir según el tipo de datos:

  - `"viridis"`: mapa perceptualmente uniforme, pasa del morado al amarillo. Es el **mapa por defecto recomendado** en Matplotlib. Ideal para datos continuos.  
  - `"plasma"`: colores brillantes, del púrpura al amarillo intenso. Bueno para resaltar contrastes.  
  - `"inferno"`: del negro al naranja y amarillo, útil en datos que representan calor o energía.  
  - `"magma"`: del negro al rosa pálido, más suave que *inferno*.  
  - `"cividis"`: diseñado para ser accesible a personas con daltonismo (buena opción inclusiva).  
  - `"hot"`: pasa de negro → rojo → amarillo → blanco, típico en mapas térmicos.  
  - `"coolwarm"`: va de azul a rojo, útil para datos centrados en un valor medio (positivo vs negativo).  
  - `"Greys"`: escala de grises, útil para resaltar intensidad sin color.

- **`aspect`** *(opcional)*  
  Controla la relación ancho/alto del gráfico.  
  - `"auto"` ajusta automáticamente.  
  - `"equal"` fuerza que cada celda sea cuadrada.

- **`interpolation`** *(opcional)*  
  Método de suavizado entre valores.  
  - `"nearest"` mantiene bordes nítidos.  
  - `"bilinear"` suaviza la transición de colores.

---

## 🔹 Escala de colores

Para agregar la barra lateral que muestra qué color corresponde a qué valor, usamos:

plt.colorbar()

## 🔹 Sintaxis

```python
plt.imshow(X, cmap=None, aspect=None, interpolation=None)



````python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. Definir el vector de probabilidades
# ==============================
# Supongamos que lanzamos una moneda "no perfectamente justa".
# La probabilidad de Sello = 0.3
# La probabilidad de Cara  = 0.7
# Lo representamos como un vector:
# p = [P(Sello), P(Cara)]
p = np.array([0.3, 0.7])

# ==============================
# 2. Convertir el vector a una "matriz fila"
# ==============================
# imshow espera una matriz 2D para poder graficar.
# reshape(1, -1) convierte [0.3, 0.7] → [[0.3, 0.7]]
# El "1" indica una fila, y "-1" deja que numpy infiera el número de columnas.
p_matriz = p.reshape(1, -1)

# ==============================
# 3. Crear la figura con varios subplots
# ==============================
# Creamos una figura con 1 fila y 3 columnas de ejes (axs es un arreglo con 3 objetos Axes).
# figsize=(12,4) define el tamaño total de la figura.
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# ==============================
# 4. Primer heatmap con el mapa de colores "viridis"
# ==============================
# Dibujamos la matriz como un heatmap en el primer eje (axs[0]).
# cmap="viridis" usa un degradado verde-azul-amarillo.
im1 = axs[0].imshow(p_matriz, cmap="viridis", aspect="auto")
axs[0].set_title("cmap = viridis")       # Título del subplot
plt.colorbar(im1, ax=axs[0])             # Barra de color asociada al heatmap

# ==============================
# 5. Segundo heatmap con el mapa de colores "inferno"
# ==============================
# "inferno" usa un degradado oscuro → rojo → amarillo.
im2 = axs[1].imshow(p_matriz, cmap="inferno", aspect="auto")
axs[1].set_title("cmap = inferno")
plt.colorbar(im2, ax=axs[1])

# ==============================
# 6. Tercer heatmap con el mapa de colores "cividis"
# ==============================
# "cividis" fue diseñado para ser perceptible por personas con daltonismo.
im3 = axs[2].imshow(p_matriz, cmap="cividis", aspect="auto")
axs[2].set_title("cmap = cividis")
plt.colorbar(im3, ax=axs[2])

# ==============================
# 7. Ajustar el espaciado entre subplots
# ==============================
# Evita que los títulos o etiquetas se sobrepongan entre sí.
plt.tight_layout()

# ==============================
# 8. Mostrar la figura completa
# ==============================
plt.show()
````

````python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. Definir la matriz de transición
# ==============================
# La moneda tiene dos estados: Sello (0) y Cara (1).
# Cada fila = estado actual.
# Cada columna = estado siguiente.
# La probabilidad de pasar de un estado a otro es 0.5 en todos los casos.
matriz_transicion = np.array([
    [0.1, 0.9],   # Fila 0: Si está en Sello → 50% Sello, 50% Cara
    [0.9, 0.2]    # Fila 1: Si está en Cara  → 50% Sello, 50% Cara
])

# ==============================
# 2. Crear la figura y el eje
# ==============================
# plt.subplots devuelve una tupla con dos objetos:
#   - El "lienzo" (Figure), donde se colocan los ejes.
#   - El "sistema de coordenadas" (Axes), donde se dibuja.
figura, eje = plt.subplots(figsize=(5, 4))  
# figsize=(5,4) define el tamaño de la figura en pulgadas (ancho x alto).

# ==============================
# 3. Dibujar la matriz como imagen (heatmap)
# ==============================
# imshow transforma la matriz numérica en una representación gráfica de colores.
imagen = eje.imshow(
    matriz_transicion,  # la matriz que queremos visualizar
    cmap="Blues",       # esquema de colores
    aspect="equal"      # mantiene las celdas cuadradas
)

# ==============================
# 4. Configurar los ejes
# ==============================
# Definimos las etiquetas que representan los estados de la moneda.
etiquetas_estados = ["Sello", "Cara"]

# Posiciones de los ticks (coinciden con los índices de la matriz: 0 y 1).
eje.set_xticks([0, 1])  
eje.set_yticks([0, 1])  

# Asignamos los nombres de estados a esas posiciones.
eje.set_xticklabels(etiquetas_estados)  
eje.set_yticklabels(etiquetas_estados)  

# ==============================
# 5. Títulos y rótulos
# ==============================
eje.set_title("Matriz de transición de la moneda")  # título general
eje.set_xlabel("Estado siguiente")                  # eje X = columnas
eje.set_ylabel("Estado actual")                     # eje Y = filas

# ==============================
# 6. Barra de colores (leyenda)
# ==============================
# La barra de color explica cómo los valores de la matriz se traducen en colores.
plt.colorbar(imagen, ax=eje)

# ==============================
# 7. Mostrar el gráfico
# ==============================
plt.show()
````

# 📊 Visualización de frecuencias con `plt.hist`

## 🔹 Objetivo
Un **histograma** es una representación gráfica que muestra la **frecuencia** con la que ocurren ciertos valores dentro de un conjunto de datos.  
En lugar de mostrar cada valor individualmente, los agrupa en **intervalos (bins)** y dibuja barras para representar cuántos datos caen en cada uno.  

Por ejemplo, si simulamos **50 sorteos de lotería** con números de 4 cifras (0000–9999), podemos analizar la **frecuencia de aparición de las cifras finales (0–9)** en esos sorteos.  

De esta manera, tenemos un vector de frecuencias que describe la distribución empírica de los dígitos:  

$$
\vec{f} =
\begin{bmatrix}
f(0) & f(1) & f(2) & \dots & f(9)
\end{bmatrix}
$$

donde $f(i)$ es la frecuencia del dígito $i$ en las unidades de los sorteos.

---

## 🔹 Parámetros principales

- **`x`**  
  Lista, array o conjunto de datos numéricos que se desean analizar.  
  - Ejemplo: las últimas cifras de los 50 números sorteados.  

- **`bins`**  
  Número de intervalos (o bordes de clase) en los que se agrupan los datos.  
  - Ejemplo: `bins=10` para separar los números en 10 grupos (0–9).  

- **`density`** *(opcional)*  
  - `False` (por defecto): el histograma muestra frecuencias absolutas.  
  - `True`: el histograma muestra frecuencias relativas (probabilidades), normalizando el área total a 1.  

- **`color`** *(opcional)*  
  Define el color de las barras. Ejemplo: `"skyblue"`, `"lightgreen"`, `"salmon"`.  

- **`edgecolor`** *(opcional)*  
  Define el color de los bordes de las barras. Útil para mejorar la visibilidad.  

- **`alpha`** *(opcional)*  
  Controla la transparencia de las barras (`0` totalmente transparente, `1` totalmente opaco).  

---

## 🔹 Sintaxis de `plt.hist`

```python
plt.hist(x, bins=None, density=False, color=None, edgecolor=None, alpha=None)


````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 🎲 1. Generar datos simulados de la lotería
# ----------------------------------------------------
# Vamos a simular 50 sorteos de un número de 4 cifras (0000 a 9999).
# Usamos randint para obtener números enteros en ese rango.
sorteos = np.random.randint(0, 10000, size=50)

# Extraemos la última cifra (unidades) de cada número.
# Esto se hace con el operador módulo % 10.
ultimas_cifras = sorteos % 10

# ----------------------------------------------------
# 📊 2. Graficar el histograma
# ----------------------------------------------------
plt.figure(figsize=(8, 5))  # Definimos tamaño de la figura

# Creamos el histograma
plt.hist(
    ultimas_cifras,     # Datos a graficar
    bins=10,            # 10 intervalos (0 al 9)
    color="skyblue",    # Color de las barras
    edgecolor="black",  # Color de los bordes de las barras
    alpha=0.7           # Transparencia de las barras
)

# ----------------------------------------------------
# 📝 3. Personalizar el gráfico
# ----------------------------------------------------
plt.title("Distribución de las últimas cifras en 50 sorteos de lotería")
plt.xlabel("Cifra final (0–9)")
plt.ylabel("Frecuencia")
plt.xticks(range(10))  # Etiquetas del eje X (0 a 9)

# Mostrar gráfico
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 🎲 1. Simulación de los sorteos
# ----------------------------------------------------
# Generamos 50 números enteros entre 0 y 9999 (inclusive).
# Esto simula 50 sorteos de lotería de 4 cifras.
sorteos = np.random.randint(0, 10000, size=50)

# Convertimos cada número entero en un string de 4 dígitos.
# Usamos f"{num:04d}" para que siempre tenga 4 posiciones:
#  - el número 42 se convierte en "0042"
#  - el número 7 en "0007"
# Esto facilita extraer cada cifra por separado.
sorteos_str = [f"{num:04d}" for num in sorteos]

# Extraemos cada dígito por posición:
#  - miles: carácter en posición [0] del string
#  - centenas: carácter en posición [1]
#  - decenas: carácter en posición [2]
#  - unidades: carácter en posición [3]
# Convertimos cada carácter a entero con int()
miles    = [int(num[0]) for num in sorteos_str]
centenas = [int(num[1]) for num in sorteos_str]
decenas  = [int(num[2]) for num in sorteos_str]
unidades = [int(num[3]) for num in sorteos_str]

# ----------------------------------------------------
# 📊 2. Cuatro histogramas separados (uno por dígito)
# ----------------------------------------------------
# Creamos una figura con 2 filas y 2 columnas de subgráficos (axs será una matriz 2x2)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# --- Histograma de los miles ---
axs[0, 0].hist(miles, bins=10, color="skyblue", edgecolor="black")
# bins=10 -> intervalos para cada dígito (0 al 9)
# color -> color de relleno de las barras
# edgecolor -> borde negro para mayor contraste
axs[0, 0].set_title("Distribución de los miles")
axs[0, 0].set_xticks(range(10))  # Mostramos los dígitos 0–9 en el eje X

# --- Histograma de las centenas ---
axs[0, 1].hist(centenas, bins=10, color="salmon", edgecolor="black")
axs[0, 1].set_title("Distribución de las centenas")
axs[0, 1].set_xticks(range(10))

# --- Histograma de las decenas ---
axs[1, 0].hist(decenas, bins=10, color="lightgreen", edgecolor="black")
axs[1, 0].set_title("Distribución de las decenas")
axs[1, 0].set_xticks(range(10))

# --- Histograma de las unidades ---
axs[1, 1].hist(unidades, bins=10, color="orange", edgecolor="black")
axs[1, 1].set_title("Distribución de las unidades")
axs[1, 1].set_xticks(range(10))

# Título general de la figura
plt.suptitle("Distribución de cada dígito en los 50 sorteos", fontsize=14)

# Ajustamos los espacios entre subplots para que no se encimen
plt.tight_layout()

# Mostramos la figura en pantalla
plt.show()

# ----------------------------------------------------
# 📊 3. Gráfico combinado con 4 series (barras lado a lado)
# ----------------------------------------------------
plt.figure(figsize=(10, 6))  # Figura nueva

# Calculamos las frecuencias absolutas de cada dígito en cada posición
# np.histogram devuelve dos cosas:
#   - counts: cuántos elementos caen en cada bin (intervalos)
#   - edges: los bordes de los intervalos (en este conteo no se usan)
fre_miles, _    = np.histogram(miles, bins=10, range=(0, 10))
fre_centenas, _ = np.histogram(centenas, bins=10, range=(0, 10))
fre_decenas, _  = np.histogram(decenas, bins=10, range=(0, 10))
fre_unidades, _ = np.histogram(unidades, bins=10, range=(0, 10))

# Creamos un vector con las posiciones del eje X: 0, 1, 2, ..., 9
x = np.arange(10)

# Definimos el ancho de cada barra
width = 0.2

# Dibujamos las 4 series de barras:
#   - Para que no se encimen, desplazamos cada grupo en el eje X
#     (ejemplo: miles va en x-1.5*width, centenas en x-0.5*width, etc.)
plt.bar(x - 1.5*width, fre_miles, width, label="Miles", color="skyblue", edgecolor="black")
plt.bar(x - 0.5*width, fre_centenas, width, label="Centenas", color="salmon", edgecolor="black")
plt.bar(x + 0.5*width, fre_decenas, width, label="Decenas", color="lightgreen", edgecolor="black")
plt.bar(x + 1.5*width, fre_unidades, width, label="Unidades", color="orange", edgecolor="black")

# Personalizamos el gráfico
plt.xticks(x, range(10))  # Etiquetas 0–9 en eje X
plt.xlabel("Dígito (0–9)")
plt.ylabel("Frecuencia")
plt.title("Distribución comparativa de cada posición en los sorteos")
plt.legend()  # Muestra la leyenda con "Miles", "Centenas", etc.

# Mostramos en pantalla
plt.show()
````

# 🏹 La función `plt.quiver` en Matplotlib

## 🔹 Descripción
`plt.quiver` permite representar **vectores como flechas** en un plano cartesiano.  
Es muy utilizada para mostrar campos vectoriales, direcciones y magnitudes.

---

## 🔹 Parámetros principales

- **`X`, `Y`**  
  Coordenadas donde comienzan las flechas.  
  Se pueden dar como escalares, listas o matrices.

- **`U`, `V`**  
  Componentes de cada flecha:  
  - `U` → desplazamiento en el eje **X** (horizontal).  
  - `V` → desplazamiento en el eje **Y** (vertical).  

- **`angles`** *(opcional)*  
  Define cómo se interpretan los ángulos:  
  - `"xy"`: coordenadas cartesianas (lo más común).  
  - `"uv"`: interpreta directamente `U, V` como dirección.  

- **`scale`** *(opcional)*  
  Factor de escala para la longitud de las flechas.  
  Si es `1`, las flechas mantienen las magnitudes originales.  

- **`scale_units`** *(opcional)*  
  Define las unidades usadas para la escala: `"xy"`, `"width"`, `"height"`, etc.  

- **`color`** *(opcional)*  
  Color de las flechas. Puede ser un solo valor o una lista.  

- **`width`** *(opcional)*  
  Grosor de las flechas en la figura.

````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 1. Definición del estado cuántico de la moneda
# ----------------------------------------------------
# Amplitudes complejas:
# Definimos un estado |ψ> = (1/√2) * [1, i]
# α = 1/√2   (para "Sello")
# β = i/√2   (para "Cara")
alpha = 1/np.sqrt(2) * 1       # amplitud para "Sello"
beta  = 1/np.sqrt(2) * 1j      # amplitud para "Cara"

# Construimos el vector de estado como array de NumPy con NOMBRE PSI
psi = np.array([alpha, beta])

# ----------------------------------------------------
# 2. Cálculo de probabilidades (módulo al cuadrado)
# ----------------------------------------------------
# Regla cuántica: P(x) = |amplitud|^2
# Esto convierte amplitudes complejas en probabilidades reales
probs = np.abs(psi)**2  # → [|α|^2, |β|^2] = [0.5, 0.5]

# ----------------------------------------------------
# 3. Crear figura con subplots
# ----------------------------------------------------
# Creamos una figura de 10 x 4 pulgadas
plt.figure(figsize=(10,4))

# SUBPLOT 1 (izquierda)
# plt.subplot(filas, columnas, índice)
# (1, 2, 1) significa:
#   - 1 fila
#   - 2 columnas
#   - este es el gráfico N° 1 (el de la izquierda)
plt.subplot(1, 2, 1)
plt.bar(["Sello", "Cara"], probs, 
        color=["skyblue", "lightcoral"], 
        edgecolor="black")
plt.title("Probabilidades $|\\alpha|^2$, $|\\beta|^2$")
plt.ylabel("Probabilidad")

# SUBPLOT 2 (derecha)
# (1, 2, 2) significa:
#   - 1 fila
#   - 2 columnas
#   - este es el gráfico N° 2 (el de la derecha)
plt.subplot(1, 2, 2)


# ----------------------------------------------------
# Dibujamos el vector de "Sello" (índice 0 del estado)
# ----------------------------------------------------
amp_sello = psi[0]   # amplitud compleja para "Sello"
plt.quiver(
    0, 0,                # origen (0,0)
    amp_sello.real,      # componente en el eje X
    amp_sello.imag,      # componente en el eje Y
    angles="xy", scale_units="xy", scale=2,
    color="blue", width=0.01
)
plt.text(
    amp_sello.real + 0.05, amp_sello.imag + 0.05,
    f"Sello\n{amp_sello.real}", fontsize=12
)

# ----------------------------------------------------
# Dibujamos el vector de "Cara" (índice 1 del estado)
# ----------------------------------------------------
amp_cara = psi[1]    # amplitud compleja para "Cara"
plt.quiver(
    0, 0, 
    amp_cara.real,
    amp_cara.imag,
    angles="xy", scale_units="xy", scale=2,
    color="red", width=0.01
)
plt.text(
    amp_cara.real + 0.05, amp_cara.imag + 0.05,
     f"Cara\n{amp_sello.real}", fontsize=12
)

# ----------------------------------------------------
# Etiquetas y título
# ----------------------------------------------------
plt.title("Amplitudes complejas de la moneda")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.grid(True)



# Dibujamos ejes X e Y
plt.axhline(0, color="black", linewidth=0.5) #
plt.axvline(0, color="black", linewidth=0.5)
plt.title("Amplitudes complejas en el plano (Re, Im)")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.grid(True)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
#plt.axis("equal")  # escala igual en X e Y para no deformar los vectores

# ----------------------------------------------------
# 4. Mostrar gráfico final
# ----------------------------------------------------
plt.tight_layout()
plt.show()
````

````python
# 📦 Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Permite gráficos en 3D

# ------------------------------------------------------
# Definición de los vectores de la base canónica en R^3
# ------------------------------------------------------
# Cada vector apunta en la dirección de un eje cartesiano:
e1 = np.array([1, 0, 0])   # Vector unitario a lo largo del eje X
e2 = np.array([0, 1, 0])   # Vector unitario a lo largo del eje Y
e3 = np.array([0, 0, 1])   # Vector unitario a lo largo del eje Z

# ------------------------------------------------------
# Crear figura 3D
# ------------------------------------------------------
fig = plt.figure(figsize=(8,8))  # Hacemos la figura un poco más grande
ax = fig.add_subplot(111, projection='3d')

# ------------------------------------------------------
# Dibujar vectores de la base ortonormal
# ------------------------------------------------------
# Cada vector arranca en el origen (0,0,0) y apunta en la dirección de su eje
ax.quiver(0, 0, 0, *e1, color='r', label=r'$e_1$', linewidth=2)
ax.quiver(0, 0, 0, *e2, color='g', label=r'$e_2$', linewidth=2)
ax.quiver(0, 0, 0, *e3, color='b', label=r'$e_3$', linewidth=2)

# ------------------------------------------------------
# Dibujar planos coordenados XY, XZ y YZ
# ------------------------------------------------------
# Usamos "meshgrid" para crear una rejilla de puntos que luego se pinta como superficie

# Rango de los ejes
xx = np.linspace(0, 1, 10)
yy = np.linspace(0, 1, 10)
XX, YY = np.meshgrid(xx, yy)

# Plano XY (Z=0)
ZZ = np.zeros_like(XX)
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='gray')  # Plano Transparente

# Plano XZ (Y=0)
YY = np.zeros_like(XX)
ZZ = np.linspace(0, 1, 10)
XX, ZZ = np.meshgrid(xx, ZZ)
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='orange')

# Plano YZ (X=0)
XX = np.zeros_like(ZZ)
YY, ZZ = np.meshgrid(yy, np.linspace(0, 1, 10))
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='cyan')

# ------------------------------------------------------
# Configuración de los ejes
# ------------------------------------------------------
ax.set_xlim([-0.5, 1.2])
ax.set_ylim([-0.5, 1.2])
ax.set_zlim([-0.5, 1.2])

# Etiquetas
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# ------------------------------------------------------
# Título y leyenda
# ------------------------------------------------------
ax.set_title("Espacio ortonormal en $\\mathbb{R}^3$: vectores base y planos coordenados")
ax.legend()

# ------------------------------------------------------
# Mostrar gráfico
# ------------------------------------------------------
plt.show()
````
