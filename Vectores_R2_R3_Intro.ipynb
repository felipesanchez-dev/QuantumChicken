# Vectores en $\mathbb{R}^2$ y $\mathbb{R}^3$


## 1. Introducción

Los **vectores** son objetos matemáticos que nos permiten representar magnitudes con dirección.

En física, se utilizan para describir fuerzas, velocidades o posiciones. En matemáticas, son la base de la geometría analítica y del álgebra lineal.

Pero lo más interesante es que en **computación cuántica**, los **estados de un qubit** se representan también mediante vectores (aunque en un espacio más abstracto).

En esta sección aprenderás:
- Qué es un vector y cómo se representa en $\mathbb{R}^2$ y $\mathbb{R}^3$
- Qué operaciones básicas podemos realizar con ellos
- Cómo esta geometría nos prepara para entender la **naturaleza cuántica** de los qubits


```python

import numpy as np
import matplotlib.pyplot as plt

# Configuración gráfica para vectores 2D
def plot_vector(v, origin=[0, 0], color='r', label=''):
    plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color=color)
    if label:
        plt.text(origin[0]+v[0]*1.05, origin[1]+v[1]*1.05, label, fontsize=12)

# Ejemplo de dos vectores
v1 = np.array([2, 1])
v2 = np.array([-1, 2])

plt.figure(figsize=(6,6))
plot_vector(v1, color='blue', label='v₁')
plot_vector(v2, color='green', label='v₂')
plt.xlim(-3, 4)
plt.ylim(-3, 4)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Visualización de vectores en $\mathbb{R}^2$")
plt.show()
```

## 2. Representación de vectores en $\mathbb{R}^2$

Un vector en $\mathbb{R}^2$ se puede representar como un par ordenado de números reales:

$$
\vec{v} = (v_1, v_2)
$$

Esto significa que el vector tiene una componente horizontal (**eje x**) y una componente vertical (**eje y**).

### Representaciones comunes:
- **Geométrica**: una flecha desde el origen hasta el punto $(v_1, v_2)$.
- **Algebraica**: como fila $[v_1, v_2]$ o columna $\begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$.

Esta forma de representación es fundamental para visualizar el comportamiento de vectores y construir intuiciones geométricas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Función para graficar vectores
def plot_vector(v, origin=[0, 0], color='r', label=''):
    plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color=color)
    if label:
        plt.text(origin[0]+v[0]*1.05, origin[1]+v[1]*1.05, label, fontsize=12)

# Vectores para representar
vectors = {
    'a': np.array([1, 2]),
    'b': np.array([2, -1]),
    'c': np.array([-1.5, 1.5])
}

# Gráfico
plt.figure(figsize=(6,6))
colors = ['red', 'blue', 'green']

for (label, vec), color in zip(vectors.items(), colors):
    plot_vector(vec, color=color, label=label)

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Diferentes vectores en $\mathbb{R}^2$")
plt.show()
```

## 3. Operaciones básicas con vectores

### 📌 Suma de vectores

Dados dos vectores en $\mathbb{R}^2$:
$$
\vec{u} = (1, 2), \quad \vec{v} = (3, -1)
$$

La **suma vectorial** se define componente a componente:
$$
\vec{u} + \vec{v} = (1 + 3, 2 + (-1)) = (4, 1)
$$

---

### 📌 Producto por escalar

Si multiplicamos un vector por un número real, solo se escala su longitud:

$$
2 \cdot \vec{u} = 2 \cdot (1, 2) = (2, 4)
$$

$$
-1 \cdot \vec{v} = -1 \cdot (3, -1) = (-3, 1)
$$

---

Estas operaciones se pueden visualizar geométricamente, lo que es muy útil para desarrollar intuición.


```python
import numpy as np
import matplotlib.pyplot as plt

# Reutilizamos la función para graficar vectores
def plot_vector(v, origin=[0, 0], color='r', label=''):
    plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color=color)
    if label:
        plt.text(origin[0]+v[0]*1.05, origin[1]+v[1]*1.05, label, fontsize=12)

# Definimos los vectores
u = np.array([1, 2])
v = np.array([3, -1])
suma = u + v
doble_u = 2 * u
neg_v = -1 * v

# Etapa 1: Vectores originales u y v
plt.figure(figsize=(6,6))
plot_vector(u, color='blue', label='u = (1, 2)')
plot_vector(v, color='green', label='v = (3, -1)')
plt.title("Etapa 1: Vectores u y v")
plt.xlim(-4, 5)
plt.ylim(-3, 5)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()

# Etapa 2: Suma u + v
plt.figure(figsize=(6,6))
plot_vector(u, color='blue', label='u')
plot_vector(v, origin=u, color='green', label='v desde u')
plot_vector(suma, color='purple', label='u + v')
plt.title("Etapa 2: Suma vectorial u + v")
plt.xlim(-1, 6)
plt.ylim(-2, 6)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()

# Etapa 3: Escalado de u
plt.figure(figsize=(6,6))
plot_vector(u, color='blue', label='u')
plot_vector(doble_u, color='orange', label='2·u')
plt.title("Etapa 3: Escalado de u por 2")
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()

# Etapa 4: Escalado de v por -1
plt.figure(figsize=(6,6))
plot_vector(v, color='green', label='v')
plot_vector(neg_v, color='red', label='-1·v')
plt.title("Etapa 4: Escalado de v por -1")
plt.xlim(-4, 4)
plt.ylim(-2, 3)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```


## 4. Norma, dirección y vector unitario

### 🧮 ¿Qué es la norma de un vector?

La **norma** (o magnitud, o módulo) de un vector $\vec{v} = (v_1, v_2)$ en $\mathbb{R}^2$ es:

$
\|\vec{v}\| = \sqrt{v_1^2 + v_2^2}
$

Representa la **longitud del vector** desde el origen hasta el punto $(v_1, v_2)$.

---

### 📐 ¿Qué es un vector unitario?

Un vector es **unitario** si su norma es 1. Cualquier vector $\vec{v}$ se puede convertir en un vector unitario al **dividirlo por su norma**:

$
\hat{v} = \frac{\vec{v}}{\|\vec{v}\|}
$

---

### 📌 Ejemplo

Sea $\vec{w} = (3, 4)$. Entonces:

- Norma:  

  $ 
  \vec{w}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5 
  $ 
  
- Vector unitario:

  $ 
  \hat{w} = \left( \frac{3}{5}, \frac{4}{5} \right)
  $

El vector $\hat{w}$ apunta en la misma dirección que $\vec{w}$, pero tiene longitud 1.


```python
# Definimos el vector w
w = np.array([3, 4])
norm_w = np.linalg.norm(w)
unit_w = w / norm_w

# Imprimimos los valores
print(f"Vector w: {w}")
print(f"Norma de w: {norm_w}")
print(f"Vector unitario (ŵ): {unit_w}")

# Visualización
plt.figure(figsize=(6,6))

# Vector original
plot_vector(w, color='blue', label='w = (3,4)')

# Vector unitario (más corto pero misma dirección)
plot_vector(unit_w, color='orange', label='ŵ (unitario)')

# Configuración gráfica
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Vector y vector unitario en la misma dirección")
plt.legend()
plt.show()
```

## 5. Producto punto y ángulo entre vectores

El **producto punto** (o producto escalar) entre dos vectores $\vec{u} = (u_1, u_2)$ y $\vec{v} = (v_1, v_2)$ en $\mathbb{R}^2$ se define como:


$$
\vec{u} \cdot \vec{v} = u_1 v_1 + u_2 v_2
$$

Este producto también puede expresarse en función del ángulo $\theta$ entre los dos vectores:


$$
\vec{u} \cdot \vec{v} = \|\vec{u}\| \|\vec{v}\| \cos(\theta)
$$


---

### ✳️ Interpretación geométrica

- Si $\vec{u} \cdot \vec{v} > 0$, el ángulo entre ellos es **agudo**.
- Si $\vec{u} \cdot \vec{v} = 0$, los vectores son **ortogonales** (perpendiculares).
- Si $\vec{u} \cdot \vec{v} < 0$, el ángulo es **obtuso**.

---

### 📌 Ejemplo

Sean $\vec{a} = (2, 3)$ y $\vec{b} = (1, -1)$.

1. Producto punto:

$$
\vec{a} \cdot \vec{b} = 2 \cdot 1 + 3 \cdot (-1) = 2 - 3 = -1
$$


2. Como el producto punto es negativo, el ángulo entre los vectores es **mayor a 90°**.


```python
# Definir los vectores
a = np.array([2, 3])
b = np.array([1, -1])

# Calcular producto punto y ángulo
dot_product = np.dot(a, b)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
cos_theta = dot_product / (norm_a * norm_b)
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

# Mostrar resultados
print(f"Producto punto a·b = {dot_product}")
print(f"Ángulo entre a y b: {theta_deg:.2f} grados")

# Visualización
plt.figure(figsize=(6,6))
plot_vector(a, color='blue', label='a = (2,3)')
plot_vector(b, color='green', label='b = (1,-1)')
plt.xlim(-1, 3)
plt.ylim(-2, 4)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.gca().set_aspect('equal')
plt.title(f"Ángulo entre a y b: {theta_deg:.1f}°")
plt.legend()
plt.show()
```

## 6. Combinación lineal y dependencia lineal

---

### 📘 1. Combinación lineal

Una **combinación lineal** de vectores consiste en multiplicar cada vector por un escalar y luego sumarlos.  
Dado un conjunto de vectores $\vec{v}_1, \vec{v}_2, ..., \vec{v}_n$, una combinación lineal se escribe como:

$$
\vec{w} = \alpha_1 \vec{v}_1 + \alpha_2 \vec{v}_2 + \cdots + \alpha_n \vec{v}_n
$$

#### ✅ Ejemplos de combinación lineal

1. Sean $\vec{v}_1 = (1, 2)$ y $\vec{v}_2 = (3, -1)$, entonces:

$$
\vec{w} = 2\vec{v}_1 + \vec{v}_2 = 2(1,2) + (3,-1) = (5,3)
$$

2. Sean $\vec{a} = (0,1)$ y $\vec{b} = (1,1)$, entonces:

$$
\vec{w} = -1\cdot \vec{a} + 3\cdot \vec{b} = (0, -1) + (3,3) = (3,2)
$$

3. Sean $\vec{u} = (-1, 4)$ y $\vec{v} = (2, 0)$, entonces:

$$
\vec{w} = 0.5\vec{u} - \vec{v} = (-0.5, 2) - (2,0) = (-2.5, 2)
$$

---

### 📘 2. Dependencia lineal

Un conjunto de vectores es **linealmente dependiente** si al menos uno de ellos puede escribirse como combinación lineal de los otros.  
En caso contrario, son **linealmente independientes**.

#### ✅ Ejemplos de dependencia lineal

1. Vectores $\vec{v}_1 = (2,4)$ y $\vec{v}_2 = (1,2)$  
   - Son **dependientes**, porque: $\vec{v}_1 = 2\vec{v}_2$

2. Vectores $\vec{a} = (1, 0)$ y $\vec{b} = (0,1)$  
   - Son **independientes**, ya que no pueden expresarse uno en función del otro.

3. Vectores $\vec{u} = (3, 6)$, $\vec{v} = (-1, -2)$ y $\vec{w} = (0, 1)$  
   - $\vec{u}$ y $\vec{v}$ son dependientes ($\vec{u} = -3\vec{v}$), pero $\vec{w}$ rompe la dependencia ⇒ el conjunto completo es **independiente**.


```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vector(v, color='blue', label=None):
    plt.quiver(0, 0, v[0], v[1],
            angles='xy', scale_units='xy', scale=1,
            color=color, label=label, width=0.008)

v1 = np.array([2, 4])
v2 = np.array([1, 2])

plt.figure(figsize=(6, 6))

plot_vector(v1, 'orange', 'v1 = (2, 4)')
plot_vector(v2, 'blue', 'v2 = (1, 2)')

plt.xlim(-1, 3)
plt.ylim(-1, 5)
plt.title("Demostración: Dependencia lineal (v1 = 2·v2)")
plt.grid(True)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.gca().set_aspect('equal')
plt.legend()
plt.show()
```

## Base Canónica en $( \mathbb{R}^2$ y $\mathbb{R}^3)$

La **base canónica** es el conjunto de vectores que señalan la dirección de cada eje en un espacio cartesiano.

- En $\mathbb{R}^2$, la base canónica está formada por:
  $$
  \vec{e}_1 = (1, 0), \quad \vec{e}_2 = (0, 1)
  $$
- En $\mathbb{R}^3$, se incluye un tercer vector:
  $$
  \vec{e}_1 = (1, 0, 0), \quad \vec{e}_2 = (0, 1, 0), \quad \vec{e}_3 = (0, 0, 1)
  $$

Cada vector en estos espacios puede expresarse como combinación lineal de los vectores de la base canónica.


```python
import matplotlib.pyplot as plt

v = 4 * e1 + 2 * e2
print("v = 4·e1 + 2·e2 =", v)


def plot_vector(v, color='black', label=None):
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
               color=color, label=label, width=0.008)

plt.figure(figsize=(6, 6))

plot_vector(e1, 'red', 'e1 = (1,0)')
plot_vector(e2, 'green', 'e2 = (0,1)')
plot_vector(v, 'blue', 'v = 3·e1 + 2·e2')

plt.xlim(-1, 5)
plt.ylim(-1, 4)
plt.grid(True)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.gca().set_aspect('equal')
plt.title("Base Canónica en {R}2 y vector expresado en ella")
plt.legend()
plt.show()


```

### 🧠 Representación de vectores en la base canónica de $(\mathbb{R}^2)$

La **base canónica** de $(\mathbb{R}^2)$ está formada por los vectores:

- $ \vec{e}_1 = (1, 0) $
- $ \vec{e}_2 = (0, 1) $

Dado un vector cualquiera $ \vec{v} = (a, b) $, se puede escribir como:

$ \vec{v} = a \cdot \vec{e}_1 + b \cdot \vec{e}_2 $

---

### ✅ Ejemplo demostrativo

Sea el vector $ \vec{v} = (3, -2) $.

Podemos escribirlo como:

$ \vec{v} = 3 \cdot \vec{e}_1 + (-2) \cdot \vec{e}_2 $

Entonces:

- $ \vec{e}_1 = (1, 0) \Rightarrow 3 \cdot \vec{e}_1 = (3, 0) $
- $ \vec{e}_2 = (0, 1) \Rightarrow (-2) \cdot \vec{e}_2 = (0, -2) $
- Sumando: $ (3, 0) + (0, -2) = (3, -2) $

Por tanto:

$ (3, -2) = 3 \cdot (1, 0) + (-2) \cdot (0, 1) $

Esto demuestra que $ \vec{v} $ pertenece a la combinación lineal de $ \{ \vec{e}_1, \vec{e}_2 \} $, y por lo tanto, está representado en la **base canónica**.


### 🧠 Representación de vectores en la base canónica de  $\mathbb{R}^3$

La **base canónica** de $mathbb{R}^3$ está formada por los vectores:

- $ \vec{e}_1 = (1, 0, 0) $
- $ \vec{e}_2 = (0, 1, 0) $
- $ \vec{e}_3 = (0, 0, 1) $

Dado un vector cualquiera $ \vec{v} = (a, b, c) $, se puede escribir como:

$ \vec{v} = a \cdot \vec{e}_1 + b \cdot \vec{e}_2 + c \cdot \vec{e}_3 $

---

### ✅ Ejemplo demostrativo

Sea el vector $ \vec{v} = (4, -1, 2) $.

Podemos escribirlo como:

$ \vec{v} = 4 \cdot \vec{e}_1 + (-1) \cdot \vec{e}_2 + 2 \cdot \vec{e}_3 $

Entonces:

- $ \vec{e}_1 = (1, 0, 0) \Rightarrow 4 \cdot \vec{e}_1 = (4, 0, 0) $
- $ \vec{e}_2 = (0, 1, 0) \Rightarrow (-1) \cdot \vec{e}_2 = (0, -1, 0) $
- $ \vec{e}_3 = (0, 0, 1) \Rightarrow 2 \cdot \vec{e}_3 = (0, 0, 2) $

Sumando:  
$ (4, 0, 0) + (0, -1, 0) + (0, 0, 2) = (4, -1, 2) $

Por tanto:

$ (4, -1, 2) = 4 \cdot (1, 0, 0) + (-1) \cdot (0, 1, 0) + 2 \cdot (0, 0, 1) $

Esto demuestra que $ \vec{v} $ está representado en la **base canónica** de $\mathbb{R}^3$.


## 🔁 ¿Cómo describimos un vector?

### 🧱 Desde la base canónica en $ \mathbb{R}^2 $

La base canónica está formada por los vectores unitarios:

- $ \hat{i} = (1, 0) $
- $ \hat{j} = (0, 1) $

---

### 🧮 Representación de cualquier vector

Cualquier vector $ \vec{v} = (a, b) \in \mathbb{R}^2 $ se puede escribir como:

$ \vec{v} = a \cdot \hat{i} + b \cdot \hat{j} $

Esto significa que multiplicamos $ \hat{i} $ por la componente $ a $, y $ \hat{j} $ por la componente $ b $, luego los sumamos para obtener $ \vec{v} $.

---

### 🎯 Ejemplo:

Sea $ \vec{v} = (3, 2) $, entonces:

$ \vec{v} = 3 \cdot \hat{i} + 2 \cdot \hat{j} = 3 \cdot (1, 0) + 2 \cdot (0, 1) = (3, 0) + (0, 2) = (3, 2) $

---

### 🔎 ¿Por qué es importante la base?

La base permite expresar vectores en coordenadas conocidas. Esto es clave para:

- Ubicar vectores en el plano.
- Interpretar su dirección y magnitud.
- Cambiar de perspectiva mediante transformaciones.
